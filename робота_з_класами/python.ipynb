–†—ñ–∑–Ω–∏—Ü—è –º—ñ–∂ –∫—ñ–ª—å–∫—ñ—Å—Ç—é —ñ–º–µ–Ω —É —Å–ø–∏—Å–∫—É names —Ç–∞ –∑–Ω–∞—á–µ–Ω–Ω—è–º total_names –≤–∏–Ω–∏–∫–∞—î —Ç–æ–º—É, —â–æ –∑–Ω–∞—á–µ–Ω–Ω—è None –ø—Ä–∏–∑–≤–æ–¥–∏—Ç—å –¥–æ —Å—Ç–≤–æ—Ä–µ–Ω–Ω—è –¥–æ–¥–∞—Ç–∫–æ–≤–æ–≥–æ –æ–±‚Äô—î–∫—Ç–∞ Anonymous, –∞ –∑–º—ñ–Ω–Ω–∞ –∫–ª–∞—Å—É total_names –ø—ñ–¥—Ä–∞—Ö–æ–≤—É—î –≤—Å—ñ —Å—Ç–≤–æ—Ä–µ–Ω—ñ –æ–±‚Äô—î–∫—Ç–∏.

üîπ –ú–æ–¥–∏—Ñ—ñ–∫–æ–≤–∞–Ω–∏–π –∫–ª–∞—Å (–£–°–Ü –í–ò–ú–û–ì–ò –í–ò–ö–û–ù–ê–ù–Ü)

```python
class MyName:
    """–ö–ª–∞—Å –¥–ª—è –¥–µ–º–æ–Ω—Å—Ç—Ä–∞—Ü—ñ—ó –û–û–ü"""

    total_names = 0  # class variable

    def __init__(self, name=None, domain="itcollege.lviv.ua") -> None:
        # —è–∫—â–æ None ‚Äî —Å—Ç–≤–æ—Ä—é—î–º–æ Anonymous
        name = name if name is not None else self.anonymous_user().name

        # –ø–µ—Ä–µ–≤—ñ—Ä–∫–∞: –ª–∏—à–µ –ª—ñ—Ç–µ—Ä–∏
        if not name.isalpha():
            raise ValueError("–Ü–º'—è –º–æ–∂–µ –º—ñ—Å—Ç–∏—Ç–∏ –ª–∏—à–µ –ª—ñ—Ç–µ—Ä–∏!")

        # –∑ –≤–µ–ª–∏–∫–æ—ó –ª—ñ—Ç–µ—Ä–∏
        self.name = name.capitalize()

        self.domain = domain
        MyName.total_names += 1
        self.my_id = MyName.total_names

    # ---------- properties ----------

    @property
    def whoami(self) -> str:
        return f"My name is {self.name}"

    @property
    def my_email(self) -> str:
        return self.create_email()

    @property
    def full_name(self) -> str:
        return f"User #{self.my_id}: {self.name} ({self.my_email})"

    # ---------- methods ----------

    def create_email(self) -> str:
        return f"{self.name}@{self.domain}"

    def count_letters(self) -> int:
        return len(self.name)

    def save_to_file(self, filename="users.txt"):
        with open(filename, "a", encoding="utf-8") as f:
            f.write(self.full_name + "\n")

    # ---------- class / static ----------

    @classmethod
    def anonymous_user(cls):
        return cls("Anonymous")

    @staticmethod
    def say_hello(message="Hello to everyone!") -> str:
        return f"You say: {message}"
```

```python

print("–†–æ–∑–ø–æ—á–∏–Ω–∞—î–º–æ —Å—Ç–≤–æ—Ä—é–≤–∞—Ç–∏ –æ–± º—î–∫—Ç–∏!")

names = ("Bohdan", "Marta", None, "Andrii")  # ‚Üê –¢–£–¢ –¢–í–û–Ñ –Ü–ú º–Ø

all_names = {name: MyName(name) for name in names}

for name, me in all_names.items():
    print(f"""{">*<"*20}
Object: {me}
Name / ID: {me.name} / {me.my_id}
whoami: {me.whoami}
email: {me.my_email}
full_name: {me.full_name}
letters in name: {me.count_letters()}
hello: {me.say_hello("–ü—Ä–∏–≤—ñ—Ç –∑ Python!")}
{"<*>"*20}""")

    me.save_to_file()

print(f"–í—Å—å–æ–≥–æ —Å—Ç–≤–æ—Ä–µ–Ω–æ –æ–± º—î–∫—Ç—ñ–≤: {MyName.total_names}")
print(f"–ö—ñ–ª—å–∫—ñ—Å—Ç—å —ñ–º–µ–Ω —É —Å–ø–∏—Å–∫—É names: {len(names)}")
```
