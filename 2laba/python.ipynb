üîπ –¶–∏–∫–ª–∏ (for, while)

```python

letters = ["a", "b", "c"]
for i in range(len(letters)):
    print(f"–ù–∞ –ø–æ–∑–∏—Ü—ñ—ó {i} –∑–Ω–∞—Ö–æ–¥–∏—Ç—å—Å—è –±—É–∫–≤–∞ {letters[i]}")
else:
    print("–¶–∏–∫–ª –∑–∞–≤–µ—Ä—à–∏–≤—Å—è –±–µ–∑ break")


count = 0 
while count < 3:
    print("–õ—ñ—á–∏–ª—å–Ω–∏–∫:", count)
    count += 1
```

üîπ –†–æ–∑–≥–∞–ª—É–∂–µ–Ω–Ω—è (if / else)

```python
from random import randint

A = randint(0, 1)

if A == 1:
    print(f"A = {A}, —ñ—Å—Ç–∏–Ω–∞")
else:
    print(f"A = {A}, —Ö–∏–±–∞")
```

üîπ try ‚Üí except ‚Üí finally (–æ–±—Ä–æ–±–∫–∞ –ø–æ–º–∏–ª–æ–∫)

```python
A = 0

try:
    print("–†–µ–∑—É–ª—å—Ç–∞—Ç:", 10 / A)
except Exception as e:
    print("–í–∏–Ω–∏–∫–ª–∞ –ø–æ–º–∏–ª–∫–∞:", e)
finally:
    print("–ë–ª–æ–∫ finally –≤–∏–∫–æ–Ω–∞–≤—Å—è")
```

üîπ Lambda-—Ñ—É–Ω–∫—Ü—ñ—ó


(–Ø–∫ —è —Ä–æ–∑—É–º—ñ—é lambda:
Lambda ‚Äî —Ü–µ –∫–æ—Ä–æ—Ç–∫–∞ –∞–Ω–æ–Ω—ñ–º–Ω–∞ —Ñ—É–Ω–∫—Ü—ñ—è, —è–∫–∞ –≤–∏–∫–æ—Ä–∏—Å—Ç–æ–≤—É—î—Ç—å—Å—è –¥–ª—è –ø—Ä–æ—Å—Ç–∏—Ö –¥—ñ–π –±–µ–∑ —Å—Ç–≤–æ—Ä–µ–Ω–Ω—è –ø–æ–≤–Ω–æ—Ü—ñ–Ω–Ω–æ—ó —Ñ—É–Ω–∫—Ü—ñ—ó —á–µ—Ä–µ–∑ def.)

```python
def a_b_func(a, b):
    return a, b

this_is_lambda = lambda name, age: f"–ú–µ–Ω–µ –∑–≤–∞—Ç–∏ {name}, –º–µ–Ω—ñ {age} —Ä–æ–∫—ñ–≤"

print("–ó–≤–∏—á–∞–π–Ω–∞ —Ñ—É–Ω–∫—Ü—ñ—è:", a_b_func("A", 10))
print("–õ—è–º–±–¥–∞:", this_is_lambda("–ë–æ–≥–¥–∞–Ω", 18))
print(this_is_lambda(*a_b_func("–ê–Ω–¥—Ä—ñ–π", 20)))
```
